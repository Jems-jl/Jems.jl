```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "luminosities.txt"
data = np.loadtxt(path)
```

```python
logmassrange = np.arange(0.1,0.7,0.1)
Xrange = np.arange(0.1,0.9,0.1)
```

```python
cmap = plt.get_cmap('viridis')
fig, ax = plt.subplots(1,1,figsize=(5,5))
for i in range(len(data[0])):
    ax.scatter(logmassrange, np.log10(data[:,i]), label="X = " + str(round(Xrange[i],1)), color=cmap(i/len(data[0])))
ax.set_xlabel("log(Mass)")
ax.set_ylabel("log(Luminosity)")
ax.grid()
ax.legend()
```
